```python
import matplotlib.pyplot as plt
import numpy as np
from math import cos, sin, pi
```

```python
mu = 0
sigma = 1
n=500

s = np.random.normal(loc=mu, scale=sigma, size=n) # draw samples from distribution

# verify the mean and variance
print(abs(mu-np.mean(s)))
print(abs(sigma-np.std(s, ddof=1)))
```

```python
# plot histograms and probability density function (PDF)
fig, axes = plt.subplots(ncols=2, figsize=(10,5))
ax = axes.ravel()

# samples + gaussian distribution
count, bins, ignored = ax[0].hist(s, 30, density=True, label='samples')
ax[0].plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r', label='P(x)')
ax[0].set_title('Samples + Gaussian distribution')
ax[0].legend()

# log P(x) --> maximize this for MLE/MAP classification
ax[1].plot(bins, np.log(1/np.sqrt(2*pi)/sigma) - (bins-mu)**2/(2*sigma**2),
         linewidth=2, color='darkred', label='log P(x)')
ax[1].set_title('MLE/MAP classifier/maximizer')

ax[1].legend()

plt.show()
```

```python
# number of samples
n = 200
# weights of each components
weightA = 2/3
weightB = 1/3
# mean
meanA = np.array([2, 3])
meanB = np.array([2, -2])
# theta
thetaA = pi/4
thetaB = pi/6
# eigenvalues 
eigvalA = np.array([1, 2])
eigvalB = np.array([4, 1])
# eigenvectors
eigvecA1 = np.array([cos(thetaA), sin(thetaA)])
eigvecA2 = np.array([-sin(thetaA), cos(thetaA)])
eigvecB1 = np.array([cos(thetaB), sin(thetaB)])
eigvecB2 = np.array([-sin(thetaB), cos(thetaB)])

eigvecA = np.vstack((eigvecA1, eigvecA2))
eigvecB = np.vstack((eigvecB1, eigvecB2))

g_mix1 = np.array([n, weightA, meanA, eigvalA, eigvecA], dtype='object')
g_mix2 = np.array([n, weightB, meanB, eigvalB, eigvecB], dtype='object')

g_mix = np.vstack((g_mix1, g_mix2))



def generate_gaussian_mixture(gauss_mix):
    
    # initialize list for distributions
    x = []
    y = []
    
    for g in gauss_mix:
        
        # step 0 - define number of samples to generate
        num_samples = int(g[0] * g[1])
        
        # step 1 - define U
        U = np.vstack((g[4][0].T, g[4][1].T))
        
        # step 2 - define Lambda
        Lambda = np.diag(g[3])
        
        # step 3 - compute C
        C = U * Lambda * U.T
        C = C.astype('float64')
        
        # step 4 - correct for floating point negative errors
        #min_eig = np.min(np.real(np.linalg.eigvals(C)))
        #if min_eig < 0:
        #    print(min_eig)
        #    C -= 10*min_eig * np.eye(*C.shape)
            
        
        # step 4 - compute distribution
        xg, yg = np.random.multivariate_normal(g[2], C.T, num_samples).T
        
        x.append(xg)
        y.append(yg)
        
    x = np.asarray(x, dtype='object')
    y = np.asarray(y, dtype='object')
        
    return x, y

x, y = generate_gaussian_mixture(g_mix)

colors = ['red', 'blue']
labels = ['C', 'D']
markers = ['x', 'o']

plt.figure(1)

for i in range(np.shape(x)[0]):

    plt.scatter(x[i], y[i], c=colors[i], marker=markers[i], label=labels[i])

plt.suptitle('Problem 1: Generating 2D synthetic data', fontweight='bold')    
plt.title('Class 1: Gaussian mixture of C and D components')
plt.axis('equal')
plt.legend()

plt.show()
```

## Breaking down the above code to satisfy the homework requirements

1. Assume equal priors

Answer: under uniform priors, the MAP decision rule becomes the ML decision rule and the probability of error is minimized. To implement the ML rule, we compute the maximum argument of the function: log(p(x|y)) for each class.

The probability of a 1D gaussian distribution is: 

P(x) = 1/(sqrt(2*pi)*sigma)*exp(-(x-mean)^2/(2*sigma^2))

The log of this (which we are searching for the maximum argument of) is:

log(P(x)) = log(1/(sqrt(2*pi)*sigma)) - (x-mean)^2/(2*sigma^2)

NOTE: the above is for a one-dimensional Gaussian distribution

## Gaussian Naive Bayes example from Python Data Science Handbook

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)

fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
ax.set_title('Naive Bayes Model', size=14)

xlim = (-8, 8)
ylim = (-15, 5)

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red', 'blue']):
    mask = (y == label)
    
    mu, std = X[mask].mean(0), X[mask].std(0)
    
    print(np.shape(X))
    print(np.shape(y))
    print(np.shape(Xgrid))
    print(np.shape(X[mask]))
    print(np.shape(mu))
    print(np.shape(std))
    
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=color, alpha=0.2)
    
ax.set(xlim=xlim, ylim=ylim)
```
