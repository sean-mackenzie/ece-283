## Generate a Gaussian mixture model in 2D

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster, datasets, mixture
from scipy.stats import multivariate_normal
from sklearn.datasets import make_spd_matrix

plt.rcParams["axes.grid"] = False
font = {'family' : 'Tahoma',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
```

# Apply the Maximum Likelihood Decision Rule because we assume uniforms priors

```python
colors = ['lightblue', 'wheat', 'darkseagreen', 'thistle', 'yellow', 'red', 'brown', 'grey']
ccolors = ['darkblue', 'darkorange', 'lime', 'deeppink']
labels = ['Class 0, A', 'Class 0, B', 'Class 1, C', 'Class 1, D']
eps=1e-8

# define the number of samples
n = 200

# define the mean points for each of the synthetic cluster centers
mu_0a = [0,0]
mu_0b = [6,4]
mu_1c = [2,3]
mu_1d = [2,-2]

t_means = [mu_0a, mu_0b, mu_1c, mu_1d] 
# t_means shape: KxN where K is the number of components and N is number of dimensions (2)

# for each cluster center, create a positive semi definite covariance matrix
cov_0a = np.array([[4,0],[0,1]])
cov_0b = np.array([[1,0],[0,4]])
cov_1c = np.array([[0.5,0],[0,1]])
cov_1d = np.array([[3,0],[0,0.75]])
t_covs = [cov_0a, cov_0b, cov_1c, cov_1d]
# t_covs shape: KxNxN where the NxN is the covariance matrix of each component K.

# weights to generate samples
sample_weights = [3/4, 1/4, 2/3, 1/3]

XX = []
X = []
for mean, cov, w in zip(t_means, t_covs, sample_weights):
    num_samples = int(n*w)
    x = np.random.multivariate_normal(mean, cov, num_samples) # create normal gaussian distributions
    X += list(x)                                    # stack points into dataset list
    XX.append(x)                                    # stack points into columns vectors for storage

X = np.array(X)                                     # make numpy array (speed)
np.random.shuffle(X)                                # shuffle the data around

print("Dataset shape: ", X.shape)


x = np.linspace(np.min(X[...,0])-1,np.max(X[...,0])+1,100)
y = np.linspace(np.min(X[...,1])-1,np.max(X[...,1])+1,80)
X_,Y_ = np.meshgrid(x,y)

pos = np.array([X_.flatten(),Y_.flatten()]).T       # stack meshgrid points into flat columns

print("Position shape: ", pos.shape, " where pos is two columns vectors of points x and y")

# define the number of clusters to be learned
k = 4

# create and initialize the cluster centers and the weight paramters
weights = np.ones((k)) / k                          # this assumes equal priors

# For a given mean
means = t_means

print("Means: ", means)
print("Weights: ", weights)

# create and initialize a Positive semidefinite convariance matrix 
# Here, we know the covariance matrix already
cov = np.array(t_covs)
print(cov.shape)

# setup our figure
fig, ax = plt.subplots(figsize=(10,8))

# STEP 2: for each class, compute the likelihood by evaluating the probability density function
likelihood = []
probability = []
for j in range(k):
    
    # Compute the log probability density function for each class
    likelihood.append(np.log(multivariate_normal.pdf(x=pos, mean=means[j], cov=cov[j])))
    
    # Compute the probability density function for each class
    probability.append(multivariate_normal.pdf(x=pos, mean=means[j], cov=cov[j]))
    
likelihood = np.array(likelihood)
probability = np.array(probability)
    
# STEP 3: Apply the Maximum Likelihood Decision Rule to determine most probable class.
predictions = np.argmax(likelihood, axis=0)
probabilities = np.max(probability, axis=0)

# STEP 4: Scatter plot the highest likelihood class with the corresponding color (decision boundary) 
for c in range(k):

    pred_ids = np.where(predictions == c)
    plt.scatter(pos[pred_ids[0],0], pos[pred_ids[0],1], color=colors[c], alpha=0.2, edgecolors='none', marker='s')

for c in range(k):
    plt.scatter(XX[c][:,0], XX[c][:,1], color=ccolors[c], s=100, marker='o', alpha=0.5, label=labels[c])

    
plt.xlabel(r'$\lambda_{1}$')
plt.ylabel(r'$\lambda_{2}$')
        
plt.legend(fontsize=14)

plt.show()
```

```python
print("Likelihood shape: ", np.shape(likelihood))
print("Likelihood of (k=4) for point #1: ", likelihood[:,0])
print("Prediction for point #1 is the argmax of likelihood: ", predictions[0])

print("---")

print("Probability shape: ", np.shape(probability))
print("Max probabilities shape: ", np.shape(probabilities))
print("Mean of max probabilities: ", np.mean(probabilities))
print("---")
print("number of samples to simulate: n=10*1/mean(prob)=", 10 * 1/np.mean(probabilities))
```

This analysis shows that ~700 is the minimum samples necessary to adequately estimate the conditional probability of incorrect classifcation for each class using the MAP decision rule.

# Language code for error estimation

1. for each position (x,y) in ground truth (x,y, len=4000):
2. find the nearest (x,y) grid coordinates from pos (x, y, len=8000):
3. get the index of pos at the nearest (x,y) grid coordinates
4. get the predicted class for that grid coordinates by indexing prediction (len=8000)
5. compare the prediction to the actual class for the (x,y) ground truth data point. 
6. if the prediction was incorrect, increment an error counter for that class.
